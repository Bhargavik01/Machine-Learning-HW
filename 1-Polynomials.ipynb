```python
import matplotlib.pyplot as plt 
plt.style.use('seaborn-whitegrid') 
import numpy as np 
noise_scale = 100 
number_of_samples = 50 
x =  25*(np.random.rand(number_of_samples, 1) - 0.5)


# y=5x+15X^2+2x^3 M=3
y = (5 * x) + (15 * x**2) + (2 * x**3) + noise_scale*np.random.randn(number_of_samples, 1) 

plt.plot(x,y,'ro') 
```

```python
import math 
import array
m = [1,2,3,4,5,6,7,8]
#mean square error will be appened to the below list for every value of M
MSE=[]
for ele in m:
    fit = np.poly1d(np.polyfit(x, y, ele))
    error_square=0
    for i in range(0,number_of_samples):
        diff=(y[i]-fit(x[i]))**2
        error_square=error_square+diff
        
    #mean of errors
    mean_error=error_square/number_of_samples
    MSE.append(mean_error)

print(MSE)



    
```
